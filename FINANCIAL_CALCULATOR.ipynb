```python
#CALCULATING NPV:

# Get user input for discount rate, initial investment, and number of years
rate = float(input("Enter discount rate : ")) / 100
investment = float(input("Enter initial investment: "))
n = int(input("Number of years: "))

# Initialize Discounted Cash Flow (DCF)
DCF = 0
# Loop through each year to get cash flows and calculate DCF
for t in range(1, n + 1):
    cf = float(input(f"Cash flow for year {t}: "))
    DCF += cf / (1 + rate) ** t

# Calculate Net Present Value (NPV)
npv = DCF - investment

# Print the calculated NPV and a decision based on its value
print(f"\nNPV = {npv:.2f}")
if npv > 0:
    print("Accept the project!")
else:
    print("Reject the project.")
```

```python
def calculate_irr(cash_flows, iterations=1000):
    # We start with a guess of 0.1 (10%)
    rate = 0.1
    # 'step' is how much we adjust the rate each time we guess
    step = 0.001

    for i in range(iterations):
        npv = 0
        # Calculate NPV for the current rate
        # Equation: CF / (1 + r)^t
        for t, cash_flow in enumerate(cash_flows):
            npv += cash_flow / (1 + rate)**t

        # If NPV is close enough to zero, we stop!
        if abs(npv) < 0.01:
            return rate

        # If NPV is positive, the rate needs to be higher
        if npv > 0:
            rate += step
        # If NPV is negative, the rate needs to be lower
        else:
            rate -= step

    return rate


# Get user input for cash flows
my_cash_flows = []
initial_investment = float(input("Enter the initial investment (outflow): "))
my_cash_flows.append(-initial_investment)

num_periods = int(input("Enter the number of subsequent periods for cash inflows: "))

for i in range(1, num_periods + 1):
    inflow = float(input(f"Enter cash inflow for period {i}: "))
    my_cash_flows.append(inflow)

result = calculate_irr(my_cash_flows)

print(f"The calculated cash flows are: {my_cash_flows}")
print(f"The IRR is: {result * 100:.2f}%")
```
